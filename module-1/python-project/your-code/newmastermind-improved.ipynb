```python
import random as rd
class color:
    PURPLE = '\033[95m'
    BLUE = '\033[34m'
    GREEN = '\033[32m'
    YELLOW = '\033[33m'
    PINK = "\033[35m"
    RED = '\033[91m'
    BOLD = '\033[1m'
    REVERSE = "\033[7m"
    CYAN = '\033[96m'
    END = '\033[0m'
    WHITE = "\033[97m"
    ORANGE = '\033[31m'



colors = ['red','blue','yellow','green','white','pink','purple','orange']
player_total_score = 0
machine_total_score = 0
number_of_trys = 12
correct_color = 0
correct = 0



def machinechoice():
    machine_choice = rd.sample(colors, 4)
    return machine_choice
        
    
def playerchoice():
    player_choice = []
    for i in range(4):
        if i ==0:
            player_color = str(input('\n''Please choose a color:''\n')).lower()
        else :
            player_color = str(input('\n''Another one please:''\n')).lower()
        while player_color not in colors :
            if( not player_choice) or player_color=='exit' or player_color=='quit' or player_color=='q':
                print('Bye', player, ', Hope to see you again !')
                return None
            else:
                print(color.BOLD+'\nYour choice is not in the list... '+color.END)
                player_color = str(input('Please choose a color: ''\n')).lower()
        player_choice.append(player_color)
    return player_choice

def play():
    global player_total_score
    global machine_total_score
    machine_choice = machinechoice()
    print("\nLet's Gooooo", player,'!\n',machine_choice)
    for i in range(number_of_trys):
        machine_choice_copy = machine_choice.copy()
        player_choice = playerchoice()
        if player_choice == None:
            return None
        elif machine_choice == player_choice :
            player_total_score+=1
            print('\n''Congratulations', player,'!' '\n' '\n''You found the secret code in', i+1 ,'try'*(i==0)+'tries'*(i>0),
                  color.BOLD +'\n\nScore :'+color.END,'\nMachine :',machine_total_score,'/ Player :',player_total_score,
                  '\n----------------------------------------------')
            if playagain() == None:
                return None
        else :
            for j in range(len(machine_choice)):
                if player_choice[j] in machine_choice_copy :
                    if player_choice[j]==machine_choice[j]:
                        correct+=1
                        machine_choice_copy[machine_choice.index(player_choice[j])] = '*'
                    else:
                        correct_color+=1
                        machine_choice_copy[machine_choice.index(player_choice[j])] = '*'
            print('\nCorrect color but wrong place:', correct_color, '\nCorrect place & color:', correct,
              '\n----------------------------------------------')
    if player_choice != machine_choice :
        machine_total_score +=1
        print('Too bad', player,"... You'll make it next time!\n\nThe Secret Code was:", machine_choice,
              '\nScore :\n\nMachine :',machine_total_score,'\nPlayer :',player_total_score,
              '==============================================')
    if playagain() == None :
        return None
    



def playagain():
    again = str(input('\nWanna play again ?')).lower()
    if again == 'yes' or again == 'y' :
        play()
    else :
        print('Bye', player,'\nHope to see you back sooon !')
        return None
```
