```python
table_brand=[]
for i in range(1,68):
    if i%2==0:
            t.sleep(2)
    url=f'https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Dresses?cm_sp=topnav-_-clothing-_-dresses&pn=%i&npp=60&image_view=product&dscroll=0'
    lst=BS(r.get(url,headers=headers).content).select('a>span.designer')
    table_url=[(i.text).upper() for i in lst]
    table_brand+=table_url
table_brand
```

```python
table_price=[]
for i in range(1,68):
    if i%2==0:
            t.sleep(3)
    url='https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Dresses?cm_sp=topnav-_-clothing-_-dresses&pn=%i&npp=60&image_view=product&dscroll=0'
    lst=BS(r.get(url=url,headers=headers).content).select('span.price')
    table_url=[(i.text).strip('\n\t€').replace(',','') for i in lst]
    table_price=table_price+table_url
table_price
```

```python
table_desc=[]
for i in range(1,68):
    if i%2==0:
            t.sleep(3)
    url='https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Dresses?cm_sp=topnav-_-clothing-_-dresses&pn='+str(i)+'&npp=60&image_view=product&dscroll=0'
    lst=BS(r.get(url,headers=headers).content).select('div.description>a')
    table_url=[''.join(re.findall(r'title=.*',str(j))).strip('">\"[]').replace('&amp;','&').replace('title=','').replace('"','').replace(' - NET SUSTAIN','') for j in lst]
    table_desc+=table_url     
table_desc
```

```python
def table_generator(url_pattern,headers,pages_to_scrape):
    table_brand=[]
    table_price=[]
    table_desc=[]
    for i in range(1,pages_to_scrape+1):
        if i%2==0:
            t.sleep(2)
        url=url_pattern%i
        soup=BS(r.get(url,headers=headers).content)
        lst1=soup.select('a>span.designer')
        table_url1=[(j.text).upper() for j in lst1]
        table_brand+=table_url1
        lst2=soup.select('span.price')
        table_url2=[(k.text).strip('\n\t€').replace(',','') for k in lst2]
        table_price+=table_url2
        lst3=soup.select('div.description>a')
        table_url3=[''.join(re.findall(r'title=.*',str(l))).strip('">\"[]').replace('&amp;','&').replace('title=','').replace('"','').replace(' - NET SUSTAIN','') for l in lst3]
        table_desc+=table_url3
        
        
    return table_brand, table_desc, table_price
```

```python
Brand1,Description1,OriginalPrice1=table_generator(f'https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Pants?cm_sp=topnav-_-clothing-_-pants&pn=%i&npp=60&image_view=product&dScroll=0',headers,25)
df_pants=pd.DataFrame([Brand1,Description1,OriginalPrice1], index=colnames).T
df_pants=df_pants.drop_duplicates()
df_pants.Description='Pants'
df_pants
```

```python
Brand2,Description2,OriginalPrice2=table_generator(f'https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Jackets?cm_sp=topnav-_-clothing-_-jackets&pn=%i&npp=60&image_view=product&dScroll=0',headers,19)
df_jacket=pd.DataFrame([Brand2,Description2,OriginalPrice2], index=colnames).T
df_jacket=df_jacket.drop_duplicates()
df_jacket.Description='Jacket'
df_jacket
```

```python
Brand3,Description3,OriginalPrice3=table_generator(f'https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Tops?cm_sp=topnav-_-clothing-_-tops&pn=%i&npp=60&image_view=product&dScroll=0',headers,47)
df_top=pd.DataFrame([Brand3,Description3,OriginalPrice3], index=colnames).T
df_top=df_top.drop_duplicates()
df_top.Description='Top'
df_top
```

```python
Brand4,Description4,OriginalPrice4=table_generator(f'https://www.net-a-porter.com/fr/fr/d/Shop/Clothing/Skirts?cm_sp=topnav-_-clothing-_-skirts&pn=%i&npp=60&image_view=product&dScroll=0',headers,17)
df_skirts=pd.DataFrame([Brand4,Description4,OriginalPrice4], index=colnames).T
df_skirts=df_skirts.drop_duplicates()
df_skirts.Description='Skirt'
df_skirts
```

```python
def mode_agg(x):
    return(''.join(x.mode().values.tolist()))
```

```python
df_skirts.OriginalPrice=df_skirts.OriginalPrice.astype('int')
df_skirts1=df_skirts.groupby('Brand').agg({'Description':mode_agg,'OriginalPrice':'mean'}).reset_index().copy()
df_skirts1
```

```python
df_dress.OriginalPrice=df_dress.OriginalPrice.astype('int')
df_dress1=df_dress.groupby('Brand').agg({'Description':mode_agg,'OriginalPrice':'mean'}).reset_index().copy()
df_dress1
```

```python
df_pants.OriginalPrice=df_pants.OriginalPrice.astype('int')
df_pants1=df_pants.groupby('Brand').agg({'OriginalPrice':'mean','Description':mode_agg}).reset_index().copy()
df_pants1
```

```python
df_top.OriginalPrice=df_top.OriginalPrice.astype('int')
df_top1=df_top.groupby('Brand').agg({'OriginalPrice':'mean','Description':mode_agg}).reset_index().copy()
df_top1
```

```python
df_jacket.OriginalPrice=df_jacket.OriginalPrice.astype('int')
df_jacket1=df_jacket.groupby('Brand').agg({'OriginalPrice':'mean','Description':mode_agg}).reset_index().copy()
df_jacket1
```
