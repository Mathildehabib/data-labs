```python
Brand4,Description4,SalePrice4,OriginalPrice4=WS.web_scraper('https://www.theoutnet.com/Search/RenderProductsAsync?ytosQuery=true&linkdepartment=INTL_Skirts_CLOTHING&linkdepartmentId=3074457345616677816&department=INTL_Skirts_CLOTHING&departmentId=3074457345616677816&gender=&brand=&macro=&micro=&season=&color=&size=&site=&section=&sortRule=&yurirulename=&microcolor=&agerange=&macroMarchio=&page=%i&productsPerPage=96&modelnames=&look=&washtype=&fabric=&prints=&suggestion=false&suggestionValue=&material=&occasion=&weight=&gallery=&colortype=&style=&issale=&heeltype=&wedge=&salesline=&family=&environment=&structure=&authorlocalized=&themecollection=&waist=&stone=&filter=&collection=&price=&facetsvalue=&model=&virtualnavigation=&textSearchFilters=&textSearch=&minMaxPrice=&searchType=&fabricColor=&modelFabric=&itembinding=&totalPages=20&rsiUsed=false&totalItems=1865&partialLoadedItems=96&itemsToLoadOnNextPage=96&siteCode=THEOUTNET_FR&langId=19',20)
df_skirt=pd.DataFrame([Brand4,Description4,SalePrice4,OriginalPrice4],index=colnames).T
df_skirt=df_skirt.drop_duplicates()
df_skirt.Description='Skirt'
df_skirt
```

```python
def mode_agg(x):
    return(''.join(x.mode().values.tolist()))
```

```python
df_skirt=df_skirt.drop((df_skirt[df_skirt.OriginalPrice==''].index.tolist()),axis=0)
df_skirt
```

```python
df_skirt.OriginalPrice=df_skirt.OriginalPrice.astype('int')
df_skirt.SalePrice=df_skirt.SalePrice.astype('int')
df_skirt1=df_skirt.groupby('Brand').agg({'Description':mode_agg,'SalePrice':'mean','OriginalPrice':'mean'}).reset_index().copy()
df_skirt1
```

```python
df_dress=df_dress.drop((df_dress[df_dress.OriginalPrice==''].index.tolist()),axis=0)
df_dress.OriginalPrice=df_dress.OriginalPrice.astype('int')
df_dress.SalePrice=df_dress.SalePrice.astype('int')
df_dress1=df_dress.groupby('Brand').agg({'Description':mode_agg,'SalePrice':'mean','OriginalPrice':'mean'}).reset_index().copy()
df_dress1
```

```python
df_jacket=df_jacket.drop((df_jacket[df_jacket.OriginalPrice==''].index.tolist()),axis=0)
df_jacket.OriginalPrice=df_jacket.OriginalPrice.astype('int')
df_jacket.SalePrice=df_jacket.SalePrice.astype('int')
df_jacket1=df_jacket.groupby('Brand').agg({'Description':mode_agg,'SalePrice':'mean','OriginalPrice':'mean'}).reset_index().copy()
df_jacket1
```

```python
df_top=df_top.drop((df_top[df_top.OriginalPrice==''].index.tolist()),axis=0)
df_top.OriginalPrice=df_top.OriginalPrice.astype('int')
df_top.SalePrice=df_top.SalePrice.astype('int')
df_top1=df_top.groupby('Brand').agg({'Description':mode_agg,'SalePrice':'mean','OriginalPrice':'mean'}).reset_index().copy()
df_top1
```
