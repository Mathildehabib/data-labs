```python
import requests as r
from bs4 import BeautifulSoup as BS
import pandas as pd
import re
import time as t
```

```python
table_dress=[]
for i in range(1,78):
    if i%2==0:
            t.sleep(2)
    url=f'https://www.theoutnet.com/Search/RenderProductsAsync?ytosQuery=true&linkdepartment=INTL_Dresses_CLOTHING&linkdepartmentId=3074457345616677807&department=INTL_Dresses_CLOTHING&departmentId=3074457345616677807&page={i}&productsPerPage=96&suggestion=false&facetsvalue=&totalPages=75&rsiUsed=false&totalItems=7192&partialLoadedItems=96&itemsToLoadOnNextPage=96&siteCode=THEOUTNET_FR&langId=19'
    lst=BS(r.get(url).content).select('div.description>span.designer-name,span.discounted>span.value,span.full>span.value,div.description>span.title')
    table_url=[(j.text).strip('\r\n\txa').replace('\xa0','') for j in lst]
    table_dress+=table_url
```

```python
table_dress
```

```python
table_dress_2=[]
for i in range((len(table_dress)-1)):
    if table_dress[i]!=table_dress[i+1]:
        table_dress_2.append(table_dress[i])
table_dress_2.append(table_dress[-1])
    
table_dress_3=[]
for i in range(0,(len(table_dress_2)-1)):
    if i==0:
        table_dress_3.append(table_dress_2[i])
    elif (table_dress_2[i].isdigit()) & (table_dress_2[i+1].isdigit()==False) & (table_dress_2[i-1].isdigit()==False):
        table_dress_3.append(table_dress_2[i])
        table_dress_3.append('')
    else:
        table_dress_3.append(table_dress_2[i])
table_dress_3.append(table_dress_2[-1])
```

```python
Brand=[table_dress_3[i] for i in range(0,len(table_dress_3),4)]
SalePrice=[table_dress_3[i] for i in range(2,len(table_dress_3),4)]
OriginalPrice=[table_dress_3[i] for i in range(3,len(table_dress_3),4)]
Description=[table_dress_3[i] for i in range(1,len(table_dress_3),4)]

colnames=['Brand','Description','SalePrice','OriginalPrice']
```

```python
df1=pd.DataFrame([Brand,Description,SalePrice,OriginalPrice], index=colnames).T
df1
```
