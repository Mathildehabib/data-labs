```python
categories=['t-shirt','baskets','robe','short','veste','chaussures','pantalon','jean','sandal','sweatshirt','shirt','pyjama']
colors=['blue','white','black','grey','pink','red','navy','green','rose','yellow']
```

```python
def detect_category(x):
    for i in categories:
        if i in x.lower():
            return i

def detect_color(x):
    for i in colors:
        if i in x.lower():
            return i
```

```python
print(df1.name.apply(detect_category).isna().sum())
df1['new_category']=df1.name.apply(detect_category)
df1.new_category=df1.new_category.fillna('Other')
df1.new_category.value_counts()
```
