```python
import pandas as pd
from bs4 import BeautifulSoup as BS
import requests
import time as t
```

```python
prices=[]
categories=[]
brands=[] 
sizes=[]
urls=[]
```

```python
for i in range(50):
    url=f'https://www.yoox.com/FR/shoponline?dept=clothingwomen&gender=D&page={i+1}&season=X&clientabt=SmsMultiChannel_ON%2CNewDelivery_ON%2CRecentlyViewed_ON%2CmyooxNew_ON'
    if i%5==0:
        t.sleep(3)
    soup=BS(requests.get(url).content)
    price=soup.select('a > div.price')
    category=soup.select('a > div.microcategory')
    brand=soup.select('a > div.brand')
    size=soup.select('div.colorSize > div')
    URL=soup.select('div.itemData.text-center > a')
    for j in range(len(price)):
        prices.append(price[j])
    for j in range(len(category)):
        categories.append(category[j])
    for j in range(len(brand)):
        brands.append(brand[j])
    for j in range(len(size)):
        sizes.append(size[j])
    for j in range(len(URL)):
        urls.append(URL[j])
    print(i)
```
