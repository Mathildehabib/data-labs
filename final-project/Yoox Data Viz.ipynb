```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
yoox=pd.read_csv('yoox_cleaned.csv').drop('Unnamed: 0',axis=1)
```

```python
yoox.head()
```

```python
yoox.columns
```

```python
jumper_avg=yoox[(yoox['_Jumpers']==1)|(yoox['_Turtle Necks'])].price_actual.mean()
```

```python
yoox[(yoox['_Jumpers']==1)|(yoox['_Turtle Necks'])].price_actual.max()
```

```python
blouse_avg=yoox[yoox['_Blouses']==1].price_actual.mean()
```

```python
yoox[yoox['_Blouses']==1].price_actual.max()
```

```python
pants_avg=yoox[(yoox['_Trousers']==1)|(yoox['_Culottes']==1)].price_actual.mean()
```

```python
yoox[(yoox['_Trousers']==1)|(yoox['_Culottes']==1)].price_actual.max()
```

```python
jacket_avg=yoox[(yoox['_Bomber Jackets']==1)|(yoox['_Jackets']==1)].price_actual.mean()
```

```python
yoox[(yoox['_Bomber Jackets']==1)|(yoox['_Jackets']==1)].price_actual.max()
```

```python
dress_avg=yoox[(yoox['_Bodycon Dress']==1)|(yoox['_Maxi Dress']==1)|(yoox['_Midi Dress']==1)|(yoox['_Mini Dress']==1)].price_actual.mean()
```

```python
yoox[(yoox['_Bodycon Dress']==1)|(yoox['_Maxi Dress']==1)|(yoox['_Midi Dress']==1)|(yoox['_Mini Dress']==1)].price_actual.max()
```

```python
coat_avg=yoox[(yoox['_Coats']==1)|(yoox['_Down-coat']==1)|(yoox['_Long Coats']==1)|(yoox['_Trench Coats']==1)].price_actual.mean()
```

```python
yoox[(yoox['_Coats']==1)|(yoox['_Down-coat']==1)|(yoox['_Long Coats']==1)|(yoox['_Trench Coats']==1)].price_actual.max()
```

```python
import numpy as np
plt.figure(figsize=(20,16))
# Choose the height of the bars
height = [coat_avg, jacket_avg, dress_avg, blouse_avg, jumper_avg, pants_avg]
 
# Choose the names of the bars
bars = ('Coats', 'Jackets', 'Dresses','Blouses', 'Jumpers', 'Trousers')
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height,color='#ff7276')
 
# Create names on the x-axis
plt.xticks(y_pos, bars,fontsize=18)
plt.yticks(fontsize=18)
plt.title('Average Price Per Category\n', fontsize=24)

# Show graphic
plt.show()
```
