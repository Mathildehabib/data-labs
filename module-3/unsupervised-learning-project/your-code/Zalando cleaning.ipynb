```python
import seaborn as sns
import numpy as np
plt.figure(figsize=(12,8))
height = zalando.groupby(by='new_category')['discount_flag'].mean().sort_values(ascending=False).head(10).tolist()
 
bars = zalando.groupby(by='new_category')['discount_flag'].mean().sort_values(ascending=False).head(10).index
y_pos = np.arange(len(bars))
 
plt.bar(y_pos, height,color='pink')
 
plt.xticks(y_pos, bars,rotation=45, horizontalalignment='right', fontsize=15)

plt.show();
```

```python
sns.set(style="white")
# Compute the correlation matrix

corr = zalando.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
